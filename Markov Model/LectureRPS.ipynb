```python
import numpy as np;
```

```python
t = ['R', 'P', 'S']
p_t = np.array([[2, 2, 1], [1, 1, 2], [1, 1, 0]])
op = np.array([0.6, 0.2, 0.2])
```

```python
state = 'P'
n = 20
for i in range(n):
    if state == 'P':
        pred = np.random.choice(t, p=p_t[0] / sum(p_t[0]))
        state = np.random.choice(t, p=op)
        print(pred)
        print(state)
    elif state == 'R':
        pred = np.random.choice(t, p=p_t[1] / sum(p_t[1]))
        state = np.random.choice(t, p=op)
        print(pred)
        print(state)
    else:
        pred = np.random.choice(t, p=p_t[2] / sum(p_t[2]))
        state = np.random.choice(t, p=op)
        print(pred)
        print(state)

    print("\n")
```
